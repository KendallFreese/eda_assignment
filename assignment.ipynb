**Q2.** This question uses the Airbnb data to practice making visualizations.

  1. Load the `./data/airbnb_hw.csv` data with Pandas. This provides a dataset of AirBnB rental properties for New York City.  
  2. What are are the dimensions of the data? How many observations are there? What are the variables included? Use `.head()` to examine the first few rows of data.
  3. Cross tabulate `Room Type` and `Property Type`. What patterns do you see in what kinds of rentals are available? For which kinds of properties are private rooms more common than renting the entire property?
  4. For `Price`, make a histogram, kernel density, box plot, and a statistical description of the variable. Are the data badly scaled? Are there many outliers? Use `log` to transform price into a new variable, `price_log`, and take these steps again.
  5. Make a scatterplot of `price_log` and `Beds`. Describe what you see. Use `.groupby()` to compute a desciption of `Price` conditional on/grouped by the number of beds. Describe any patterns you see in the average price and standard deviation in prices.
  6. Make a scatterplot of `price_log` and `Beds`, but color the graph by `Room Type` and `Property Type`. What patterns do you see? Compute a description of `Price` conditional on `Room Type` and `Property Type`. Which Room Type and Property Type have the highest prices on average? Which have the highest standard deviation? Does the mean or median appear to be a more reliable estimate of central tendency, and explain why?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

## Q2.1
url_airbnb = "https://raw.githubusercontent.com/KendallFreese/eda_assignment/main/data/airbnb_hw.csv"
airbnb = pd.read_csv(url_airbnb)

print("Q2.2")
rows, cols = airbnb.shape
print("Dimensions (rows, cols):", (rows, cols))
print("Number of observations (rows):", rows)
print("\nVariables (columns):")
print(list(airbnb.columns))

print("\nFirst 5 rows:")
display(airbnb.head(5))

print("Q2.3")
# Clean names/values (avoid issues from stray spaces)
airbnb.columns = [c.strip() for c in airbnb.columns]
rt = "Room Type"; pt = "Property Type"
airbnb[rt] = airbnb[rt].astype(str).str.strip()
airbnb[pt] = airbnb[pt].astype(str).str.strip()

# Crosstab: Room Type and Property Type
ct = pd.crosstab(airbnb[rt], airbnb[pt])
print("Room Type × Property Type (counts):")
print(ct.to_string())

print("Apartments dominate Entire home/apt and Private room listings with shared rooms are rare overall.")

#  Property types with more Private rooms than Entire homes?
by_prop = ct.T  # rows = property types
if {"Private room", "Entire home/apt"} <= set(by_prop.columns):
    mask = by_prop["Private room"] > by_prop["Entire home/apt"]
    more_private_than_entire = sorted(by_prop.index[mask].tolist())
    print("\nProperty types where PRIVATE ROOMS > ENTIRE HOMES:")
    print(more_private_than_entire if more_private_than_entire else "None")
else:
    print("\nExpected room-type labels not found; got:", list(by_prop.columns))

print("Q2.4")
# use exact columns from the dataset
price = pd.to_numeric(airbnb["Price"], errors="coerce")
airbnb["price_log"] = np.log(price)

# --- raw Price ---
print("Price — describe():")
print(price.describe())

# quick outlier count (1.5*IQR)
q1, q3 = price.quantile([0.25, 0.75])
iqr = q3 - q1
f_lo, f_hi = q1 - 1.5*iqr, q3 + 1.5*iqr
print("Outliers (1.5*IQR):", ((price < f_lo) | (price > f_hi)).sum())

price.plot(kind="hist", bins=50, title="Price — Histogram"); plt.xlabel("USD"); plt.show()
price.plot(kind="kde", title="Price — KDE"); plt.xlabel("USD"); plt.show()
plt.boxplot(price.dropna(), vert=False); plt.title("Price — Boxplot"); plt.xlabel("USD"); plt.show()

print("The data isn't well scaled and Price is right-skewed with high-end outliers")

# --- log(Price) ---
logp = airbnb["price_log"].replace([np.inf, -np.inf], np.nan)
print("\nlog(Price) — describe():")
print(logp.describe())

logp.plot(kind="hist", bins=50, title="log(Price) — Histogram"); plt.xlabel("log(USD)"); plt.show()
logp.plot(kind="kde", title="log(Price) — KDE"); plt.xlabel("log(USD)"); plt.show()
plt.boxplot(logp.dropna(), vert=False); plt.title("log(Price) — Boxplot"); plt.xlabel("log(USD)"); plt.show()

print("Log(Price) is a better representation of the data which is much more symmetric/compact.")

print("Q2.5")
# 1) make numeric price once and keep it
airbnb["Price_num"] = pd.to_numeric(airbnb["Price"], errors="coerce")
airbnb["price_log"] = np.log(airbnb["Price_num"])

plt.scatter(airbnb["Beds"], airbnb["price_log"], s=6, alpha=0.3)
plt.xlabel("Beds"); plt.ylabel("log(Price)"); plt.title("log(Price) vs Beds")
plt.show()

print("log(Price) generally increases with Beds, and the spread widens for larger bed counts (more variability), with a few high-price outliers still visible.")

# 2) summary of Price by Beds  ← use Price_num, not Price
by_beds = (airbnb
           .dropna(subset=["Beds", "Price_num"])
           .groupby("Beds")["Price_num"]
           .agg(count="count", mean="mean", sd="std", median="median")
           .sort_index())

print(by_beds.round(2))

print("The mean price rises with Beds and the standard deviation increases alongside it, with occasional spikes at high Beds driven by small sample sizes.")

print("Q2.6")

# --- Scatter: log(Price) vs Beds, colored by Room Type ---
df_rt = airbnb.dropna(subset=["Beds","price_log","Room Type"])
plt.figure()
for name, g in df_rt.groupby("Room Type"):
    plt.scatter(g["Beds"], g["price_log"], s=6, alpha=0.25, label=name)
plt.legend(markerscale=3)
plt.xlabel("Beds"); plt.ylabel("log(Price)")
plt.title("log(Price) vs Beds by Room Type")
plt.show()

# --- Scatter: log(Price) vs Beds, colored by Property Type (top 6 to keep readable) ---
top_pt = airbnb["Property Type"].value_counts().head(6).index
df_pt = airbnb[airbnb["Property Type"].isin(top_pt)].dropna(subset=["Beds","price_log","Property Type"])
plt.figure()
for name, g in df_pt.groupby("Property Type"):
    plt.scatter(g["Beds"], g["price_log"], s=6, alpha=0.25, label=name)
plt.legend(markerscale=3)
plt.xlabel("Beds"); plt.ylabel("log(Price)")
plt.title("log(Price) vs Beds (Top Property Types)")
plt.show()

# --- Price description conditional on Room Type × Property Type ---
by_rp = (
    airbnb.dropna(subset=["Price_num"])
    .groupby(["Room Type","Property Type"])["Price_num"]
    .agg(count="count", mean="mean", sd="std", median="median")
    .sort_values("mean", ascending=False)
)

print("\nPrice by Room Type × Property Type (top 15 by mean):")
print(by_rp.head(15).round(2))

# Highest mean and highest sd (use count>=30 to avoid tiny groups; relax if none)
eligible = by_rp[by_rp["count"] >= 30]
if eligible.empty:
    eligible = by_rp  # fall back to all groups if nothing passes the threshold

top_mean = eligible.sort_values("mean", ascending=False).head(1)
top_sd   = eligible.sort_values("sd",   ascending=False).head(1)

print("\nHighest average price (group, mean):")
print(top_mean.round(2))
print("\nHighest standard deviation (group, sd):")
print(top_sd.round(2))

print("Patterns: For the same number of beds, Entire home/apt listings are generally priced above Private room, with Shared room the lowest")
print("Higher-capacity property types (e.g., Loft, Condominium, Townhouse) cluster at higher log(prices), and the spread widens as beds increase.")
print("Highest average price: Entire home/apt — Loft (mean approximately $284.91).")
print("Highest standard deviation: Entire home/apt — Loft again (sd approximately $184.52).")
print("The median is more reliable than the mean because prices are right-skewed with outliers that pull the mean upward (e.g., Loft mean $284.91 vs median $220), so median better reflects a “typical” listing.")

```

**Q3.** This question looks at a time series of the number of active oil drilling rigs in the United States over time. The data comes from the Energy Information Agency.

1. Load `./data/drilling_rigs.csv` and examine the data. How many observations? How many variables? Are numeric variables correctly read in by Pandas, or will some variables have to be typecast/coerced? Explain clearly how these data need to be cleaned.
2. To convert the `Month` variable to an ordered datetime variable, use `df['time'] = pd.to_datetime(df['Month'], format='mixed')`.
3. Let's look at `Active Well Service Rig Count (Number of Rigs)`, which is the total number of rigs over time. Make a line plot of this time series. Describe what you see.
4. Instead of levels, we want to look at change over time. Compute the first difference of  `Active Well Service Rig Count (Number of Rigs)` and plot it over time. Describe what you see.
5. The first two columns are the number of onshore and offshore rigs, respectively. Melt these columns and plot the resulting series.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

## Q3.1
url_drillrigs = "https://raw.githubusercontent.com/KendallFreese/eda_assignment/main/data/drilling_rigs.csv"
rigs = pd.read_csv(url_drillrigs)
rigs.head()

print("Q3.1")

print("Shape (rows, cols):", rigs.shape)
print("\nColumns:", list(rigs.columns))
print("\nDtypes BEFORE cleaning:")
print(rigs.dtypes)

for c in rigs.columns:
    if c != "Month":
        rigs[c] = pd.to_numeric(rigs[c].replace({"Not Available": np.nan, "not available": np.nan, "NA": np.nan}),
                                errors="coerce")

print("\nDtypes AFTER cleaning:")
print(rigs.dtypes)

print("The file loads with ~623 rows × 10 columns")
print("Several numeric columns are read as object because of 'Not Available' strings—replace them with NaN and use pd.to_numeric(..., errors='coerce').")


print("Q3.2")

rigs["time"] = pd.to_datetime(rigs["Month"], format="mixed")
rigs = rigs.sort_values("time")
print("\nTime span:", rigs["time"].min().date(), "to", rigs["time"].max().date())


print("Q3.3")
active = "Active Well Service Rig Count (Number of Rigs)"
plt.figure()
plt.plot(rigs["time"], rigs[active])
plt.title("Active Well Service Rig Count (levels)")
plt.xlabel("Month"); plt.ylabel("Rigs")
plt.show()

print("Rigs rise steeply into the late-1970s/early-1980s, peak, then collapse in the mid-1980s and cycle through smaller booms and busts (1998–2000, 2008–09, 2014–16), with a sharp plunge around 2020 and only a partial rebound afterward.")
print("Overall, the highs and lows trend downward over time—well below the 1980s peak.")

print("Q3.4")
rigs["active_diff"] = rigs[active].diff()
plt.figure()
plt.plot(rigs["time"], rigs["active_diff"])
plt.axhline(0, linestyle="--")
plt.title("Change in Active Well Service Rig Count (first difference)")
plt.xlabel("Month"); plt.ylabel("Δ rigs")
plt.show()

print("Month-to-month changes mostly hover around zero, but there are intermittent big shocks.")
print("Volatility is highest in the 1980s and early 2000s, with a few extreme negative drops and some large positive jumps.")

print("Q3.5")
# Use the named columns to be clear:
on_col  = "Crude Oil and Natural Gas Rotary Rigs in Operation, Onshore (Number of Rigs)"
off_col = "Crude Oil and Natural Gas Rotary Rigs in Operation, Offshore (Number of Rigs)"

long = rigs[["time", on_col, off_col]].melt(
    id_vars="time", value_vars=[on_col, off_col],
    var_name="rig_type", value_name="rigs"
)

plt.figure()
for k, g in long.groupby("rig_type"):
    plt.plot(g["time"], g["rigs"], label=k)
plt.legend()
plt.title("Onshore vs Offshore Rigs in Operation")
plt.xlabel("Month"); plt.ylabel("Number of rigs")
plt.show()
```
